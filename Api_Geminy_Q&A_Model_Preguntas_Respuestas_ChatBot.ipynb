## ExtracciÃ³n de Datos

```python
## IteraciÃ³n con bucles anidados
for comentario in data['Comentarios']:
    for elemento in comentario:
        if 'commenter_name' in elemento: # or if elemento[0] == "comment_id":
            print(elemento['commenter_name'])
```

```python
## extraer nombres y 'comment_text' de data['Comentarios']  iterando
extract_names = []
for comentario in data['Comentarios']:
    for elemento in comentario:
        if 'commenter_name' in elemento:
            extract_names.append(elemento['commenter_name'])
```

```python
## ExtracciÃ³n de "commenter_name" y replicas de "commenter_name", distintas a la rÃ©plicas de Bancolombia
extract_text = []
for comentario in data['Comentarios']:
    for elemento in comentario:
        for replies in elemento['replies']:
            if 'comment_text' in replies and 'commenter_name' in replies and replies['commenter_name'] != 'Bancolombia':
                extract_text.append(replies['comment_text'])
```

```python
extract_text
```

```python
print(type(data))
print(type(data["Comentarios"]))
```

```python
## Extraer replicas de Bancolombia
## Extact 'commenter_text' from 'commenter_name' == 'Bancolombia'
bextract_text = []
for comentario in data['Comentarios']:
    for elemento in comentario:
        for replies in elemento['replies']:
            if 'comment_text' in replies and 'commenter_name' in replies and replies['commenter_name'] == 'Bancolombia':
                bextract_text.append(replies['comment_text'])
```

```python
print(len(bextract_text)) ## Output Data
print(len(extract_text)) ## Inputs data
```

```python
bextract_text
```

## Prepare the data to modeling



```python
## Generar inputs y outputs para entrenar el modelo (Chatbot)
inputs = extract_text
outputs = bextract_text
print(len(inputs))
print(len(outputs))
```

```python
## Filtrar Datos de entrada,  para que solo incluyan aquellos comentarios que tienen una respuesta de Bancolombia.
filtered_inputs = []
filtered_outputs = []

for input_text, output_text in zip(inputs, outputs):
    if output_text:  # Verifica si hay una respuesta de Bancolombia
        filtered_inputs.append(input_text)
        filtered_outputs.append(output_text)

# Ahora filtered_inputs y filtered_outputs tienen la misma longitud
```

```python
print(len(filtered_inputs))
print(len(filtered_outputs))
```

```python
print(filtered_inputs)
```

```python
print(filtered_outputs)
```

## Train the Model

```python
## Create dictionary list
training_data = [
    {"input_text": input_text, "extraction": output_text}
    for input_text, output_text in zip(filtered_inputs, filtered_outputs)
]
```

```python
type(training_data)
```

```python
## take a sample for validation data
validation_data = training_data[-35:]
training_data = training_data[:-35]
```
